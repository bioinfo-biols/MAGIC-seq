```python
import matplotlib.pyplot as plt
import pandas as pd
from math import pi
plt.figure(figsize=(5,5))
sns.set(style="ticks", font_scale=1.3)
# Set data
df = pd.DataFrame({
'group': ['MAGIC-seq','Stereo-seq','10x visum','Slide-seqv2','DBiT-seq','Decoder-seq'],
'Area': [4,4, 3, 2 , 3 , 3 ],
'Resolution': [3,4, 2, 3, 3, 3],
'Cost': [4,1, 2, 1, 3, 3.5],
'Throughput': [4,2, 3, 2, 2,2],
'Sensitivity': [3,3, 3, 2,3,3]
})
# ------- PART 1: Create background
# number of variable
categories=list(df)[1:]
N = len(categories)
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
# Initialise the spider plot
ax = plt.subplot(111, polar=True)
# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
# Draw one axe per variable + add labels
plt.xticks(angles[:-1], categories)
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([1,2,3], ["1","2","3"], color="grey", size=7)
plt.ylim(0,4)
# ------- PART 2: Add plots
# Plot each individual = each line of the data
# I don't make a loop, because plotting more than 3 groups makes the chart unreadable
# Ind1

# Ind3

values=df.loc[1].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid',color='#f7931e', label="Stereo-seq")
ax.fill(angles, values, '#f7931e', alpha=0.1)
# Ind2
values=df.loc[2].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid',color='#2171b5' ,label="10x Visium")
ax.fill(angles, values, '#e49c75', alpha=0.0)

# Ind3
values=df.loc[3].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', color='#ffeda0',label="Slide-seqv2")
ax.fill(angles, values, '#ffeda0', alpha=0.0)

# Ind3
values=df.loc[4].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', color='#dd3497', label="DBiT-seq")
ax.fill(angles, values, '#dd3497', alpha=0.0)
  
# Ind3
values=df.loc[5].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid',color= '#39b54a',label="Decoder-seq")
ax.fill(angles, values, '#39b54a', alpha=0.0)
      
values=df.loc[0].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=2, linestyle='solid',color= '#cb181d',label="MAGIC-seq")
ax.fill(angles, values, '#cb181d', alpha=0.1)

# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(1.6, 0.5))
plt.tick_params(colors='black')

plt.show()
```
