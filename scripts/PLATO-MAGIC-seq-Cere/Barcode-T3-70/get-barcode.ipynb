```python
import pandas as pd
```

```python
df_A=pd.read_csv('Spatial_barcodeA70.txt',header=None,sep='\t')
df_B=pd.read_csv('Spatial_barcodeB70.txt',header=None,sep='\t')
```

```python
Spot_coordinate={}
for a in df_A.index:
    for b in df_B.index:
        a_id=df_A.loc[a,0]
        a_seq=df_A.loc[a,1]
        b_id=df_B.loc[b,0]
        b_seq=df_B.loc[b,1]
        Spot_coordinate[a_id+'-'+b_id]=[a_seq+b_seq]

Spot_coordinate=pd.DataFrame.from_dict(Spot_coordinate,orient='index',columns=['seq'])       
```

```python
Spot_coordinate.to_csv('T3-spatial_barcode70.txt',header=None,sep='\t')
```

```python
Spot_ids=Spot_coordinate
```

```python
Spot_ids['A']=[i.split('-')[0][1:] for i in Spot_ids.index]
Spot_ids['B']=[i.split('-')[1][1:] for i in Spot_ids.index]
```

```python
Spot_ids.to_csv('T3-ids-barcode.txt',header=0,index=0,sep='\t')
```

```python
Spot_ids.iloc[:,0].to_csv('whitelist.txt',header=0,index=0,sep='\t')
```
