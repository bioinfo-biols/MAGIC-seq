```python
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import os
import re
def plot_black(ax):
    plt.tick_params(colors='black')
    ax.spines['left'].set_color('black')
    ax.spines['right'].set_color('black')
    ax.spines['bottom'].set_color('black')
    ax.spines['top'].set_color('black')
    ax.xaxis.label.set_color('black')
    ax.yaxis.label.set_color('black')
```

# Figure S2 B

```python
df_3= pd.read_csv('../file/source_data/FigureS2B.csv',index_col=0)
df_3
```

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
plt.figure(figsize=(3,6))
sns.set(style="ticks",font_scale=1.2)
# Grouped violinplot
ax=sns.boxplot(x="Label", y="Mean",
            dodge=False,
            fliersize=0,
            data=df_3)
#plt.hlines(25,colors='#e09266',linestyles='dashed')
plt.ylim(-200,4200)
plt.xlabel('')
plt.ylabel('Intensity')
#plt.hlines(25,xmin=0,xmax=2,colors='red',linestyles='dashed')
#sns.despine(offset=10, trim=True)
plot_black(ax)

plt.show()
```

# Figure 2D cy3 cy5

```python
df_cy35 = pd.read_csv('../file/source_data/Figure2D.csv',index_col=0)
df_cy35
```

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(3.5,3.5))

sns.set(style="ticks", font_scale=1)

sns.boxplot(x="reg", y="Mean", color="red", fliersize=0,data=df_cy35[df_cy35['type']=='cy3'], linewidth=0.8, width=0.5)
sns.boxplot(x="reg", y="Mean",  color="#5f9e6e",fliersize=0, data=df_cy35[df_cy35['type']=='cy5'],  linewidth=0.8,width=0.5)

plt.ylim(-2100,2700)
plt.ylabel('Intensity')

plt.show()
```

```python
df_cy3_reg9 = pd.read_csv('../file/source_data/Figure2E_S5A.csv',index_col=0)
df_cy3_reg9
```

```python
plt.figure(figsize=(3,3))
sns.kdeplot(data=df_cy3_reg9[df_cy3_reg9['Barcode']=='BarcodeB'], x="Mean", levels=100, fill=True, alpha=0.5)
sns.kdeplot(data=df_cy3_reg9[df_cy3_reg9['Barcode']=='BarcodeC'], x="Mean", levels=100, fill=True, alpha=0.5)
plt.vlines(np.median(df_cy3_reg9[df_cy3_reg9['Barcode']=='BarcodeB']['Mean']),ymin=0,ymax=0.0030,colors='#6888bc',linestyles='dashed')
plt.vlines(np.median(df_cy3_reg9[df_cy3_reg9['Barcode']=='BarcodeC']['Mean']),ymin=0,ymax=0.0030,colors='#e09266',linestyles='dashed')
plt.ylim(0,0.0028)
plt.xlim(0,3500)
plt.xlabel('Intensity')
plt.show()
```
